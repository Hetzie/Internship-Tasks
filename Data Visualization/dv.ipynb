```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
df_iris = pd.read_csv('./Iris-(1).csv')
df_iris.head()
```

```python
df_iris["Species"].value_counts()
```

```python
df_iris.drop(columns=['Id'], inplace=True)
df_iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
le = LabelEncoder()
df_iris['species'] = le.fit_transform(df_iris['species'])
# {1: 'Iris-setosa', 2: 'Iris-versicolor', 3: 'Iris-virginica'}
```

```python
df_iris.head()
```

```python
df_iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python
sns.set(style="white", color_codes=True)
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df_iris, size=4)
plt.show()
```

```python

sns.FacetGrid(df_iris, hue="Species") \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()

plt.show()
```

## Matplotlib

```python
# bar

fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (3, 3) 

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
# scatter

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.rcParams["figure.figsize"] = (3, 3) 
plt.scatter(x, y)

plt.show()
```

```python
# Line chart

plt.plot([1,2,3],[4,5,1])
plt.plot([2,2.5,4],[3,5,2])
plt.show()
```

```python
# Histogram
N_points = 100000
n_bins = 20

dist1 = np.random.standard_normal(N_points)
dist2 = 0.4 * np.random.standard_normal(N_points) + 5

plt.rcParams["figure.figsize"] = (7,4) 
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(dist1, bins=n_bins)
axs[1].hist(dist2, bins=n_bins)
```

```python
# Area

days = [1,2,3,4,5]
  
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
labels=activities,
colors=cols,
startangle=90,
shadow= True,
explode=(0,0.1,0,0),
autopct='%1.1f%%')
 
plt.title('Pie Plot')
plt.show()
```

## Seaborn

```python
# Histogram with KDE (kernel density estimate/ curve)
sns.histplot(data=df_iris, x='sepal_width', kde=True)
plt.title('Histogram with KDE')
plt.show()
```

```python
# Histogram with multiple features
sns.histplot(data=df_iris[['sepal_length','sepal_width']], palette=['red', 'blue'])
plt.title('Multi-Column Histogram')
plt.show()
```

```python
#Stacked Histogram
sns.histplot(df_iris, x='sepal_length', hue='species', multiple='stack',
             linewidth=0.5)
plt.title('Stacked Histogram')
plt.show()
```

```python
df2 = pd.read_csv('..\EDA\haberman.csv')
df2.columns = ['age','year','nodes','status']
df2.head()
```

```python
# Boxplot
plt.figure(figsize=(20,10))
for index, columnY in enumerate(list(df2.columns)[:-1]):
    plt.subplot(1, 3, index+1)
    sns.boxplot(x="status", y=columnY, data=df2, palette='husl')
plt.show()
```

```python
# violinplot
plt.figure(figsize=(20,10))
for index, columnY in enumerate(list(df2.columns)[:-1]):
    plt.subplot(1, 3, index+1)
    sns.violinplot(x="status", y=columnY, data=df2, palette='husl')
plt.show()
```

```python
# Swarmplot
plt.figure(figsize=(20,10))
for index, columnY in enumerate(list(df2.columns)[:-1]):
    plt.subplot(1, 3, index+1)
    sns.swarmplot(x="status", y=columnY, data=df2, palette='husl')
plt.show()
```

```python
#StripPlot
plt.figure(figsize=(20,10))
for index, columnY in enumerate(list(df2.columns)[:-1]):
    plt.subplot(1, 3, index+1)
    sns.stripplot(x="status", y=columnY, data=df2, palette='husl')
plt.show()
```

```python
# Hex Plot with Histogram margins
plt.figure(figsize=(20,10))
sns.jointplot(x="nodes", y='age', data=df2, palette='husl',kind='hex', ratio=2, marginal_ticks=True)
plt.show()
```

```python
# Scatter Plot with KDE Margins
plt.figure(figsize=(20,10))
sns.jointplot(x="nodes", y='age', data=df2, hue="status", height=5, ratio=2, marginal_ticks=True)
plt.show()
```
